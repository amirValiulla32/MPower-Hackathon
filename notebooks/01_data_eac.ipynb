Standardize column names: lowercase + underscores

```python
def clean_columns(df):
    df = df.copy()
    df.columns = (
        df.columns.str.strip()
                  .str.lower()
                  .str.replace(" ", "_")
    )
    return df

dataset_1_clean = clean_columns(dataset_1)
dataset_2_clean = clean_columns(dataset_2)

print("Dataset 1 columns:", dataset_1_clean.columns.tolist())
print("Dataset 2 columns:", dataset_2_clean.columns.tolist())
```

```python
dataset_2_clean = dataset_2_clean.rename(columns={'muslim' : 'muslim_score'})

print("muslim_score" in dataset_2_clean.columns)
```

```python
map_d1 = {
    "campaign_finance_supporter": "campaign_finance_support",
    "climate_change_supporter": "climate_change_support",
}
map_d2 = {
    "campaign_finance_reform": "campaign_finance_support",
    "pro_climate_change": "climate_change_support",
}

dataset_1_clean = dataset_1_clean.rename(columns={k:v for k,v in map_d1.items() if k in dataset_1_clean.columns})
dataset_2_clean = dataset_2_clean.rename(columns={k:v for k,v in map_d2.items() if k in dataset_2_clean.columns})

for df in (dataset_1_clean, dataset_2_clean):
    for col in ["campaign_finance_support", "climate_change_support"]:
        if col not in df.columns:
            df[col] = np.nan
        df[col] = pd.to_numeric(df[col], errors="coerce")

print("D1 has:", "campaign_finance_support" in dataset_1_clean.columns,
                 "climate_change_support" in dataset_1_clean.columns)
print("D2 has:", "campaign_finance_support" in dataset_2_clean.columns,
                 "climate_change_support" in dataset_2_clean.columns)
```

```python
dataset_2_clean = dataset_2_clean.drop(columns=["statefileid", "vendorid", "countyfileid","suffix", "addressid"], errors="ignore")

dataset_1_clean = dataset_1_clean.drop(columns=["statefileid", "vendorid","harris_supporter"], errors="ignore")

print("Dataset 1 columns:", dataset_1_clean.columns.tolist())
print("Dataset 2 columns:", dataset_2_clean.columns.tolist())
```

Common Values to Drop

```python
cols_to_drop = ["climate_change_support", "pro_gun_ownership", "unionsupport", 'preferred_language']

dataset_1_clean = dataset_1_clean.drop(columns=cols_to_drop, errors="ignore")
dataset_2_clean = dataset_2_clean.drop(columns=cols_to_drop, errors="ignore")

print("Dataset 1 columns:", dataset_1_clean.columns.tolist())
print("Dataset 2 columns:", dataset_2_clean.columns.tolist())
```

Merge data sets

```python
cols = dataset_1_clean.columns.tolist()
dataset_2_aligned = dataset_2_clean.reindex(columns=cols)

combined = pd.concat([dataset_1_clean, dataset_2_aligned], ignore_index=True)

combined["muslim_score"] = pd.to_numeric(combined["muslim_score"], errors="coerce")

combined = (
    combined.sort_values(["voter_file_vanid", "muslim_score"], ascending=[True, False])
            .drop_duplicates(subset=["voter_file_vanid"], keep="first")
            .reset_index(drop=True)
)

combined.info()

combined.to_csv(SAVE_CSV_PATH, index=False)
```
